### Loading the dataset

```python
import pandas as pd
import numpy as np
```

```python
bigmart = pd.read_csv('datasets\FErain_bm.csv')
```

```python
bigmart.head(10)
```

```python
data = bigmart[['Item_Visibility', 'Item_MRP','Item_Weight']]
data.head()
```

## Min Max Scaler

<img src = "min max formula.png"> 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
scaled_data = pd.DataFrame(scaled_data, columns=['Item_Visibility', 'Item_MRP','Item_Weight'])
scaled_data.head()
```

```python
scaled_data.describe()
```

## Standard Scaler

<img src = "standard scaler formula.png">

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_data = pd.DataFrame(scaled_data, columns=['Item_Visibility', 'Item_MRP'])
scaled_data.head()
```

```python
scaled_data.describe()
```
